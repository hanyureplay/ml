```python
def clean_text(s):
    s = s.lower()
    for ch in string.punctuation:                                                                                                     
        s = s.replace(ch, " ") 
    s = re.sub("[0-9]+", "||DIG||",s)
    s = re.sub(' +',' ', s)        
    return s
```

```python
train_wocha['clean_body'] = [clean_text(s) for s in train_wocha['body']]
```
